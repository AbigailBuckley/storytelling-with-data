```python
#Choropleth Map for 2014
m = fm.Map(location=[39.381266,-97.922211], zoom_start=4,tiles='Mapbox bright')

fm.Choropleth(
    geo_data=natl_geo,
    name='choropleth',
    data=data,
    columns=['State','Percentage'],
    key_on='feature.id',
    fill_color='GnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Diagnosed Diabetes, 2014'
).add_to(m)

fm.LayerControl().add_to(m)

m
```
