## Import necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load in data

```python
data = pd.read_excel('GSS 3.xlsx', sheetname='Data')
```

```python
data.head()
```

## Replace all 'not applicable'/'no issp'/'no answer' with nans

```python
data = data.replace('Not applicable', np.NaN)
data = data.replace('No issp', np.NaN)
data = data.replace('No answer', np.NaN)
```

```python
data.head()
```

## Narrow down variables

```python
variables_of_interest = ['Year of birth',
                        'Region of interview',
                        'Condition of health',
                        'General happiness',
                        'How much time felt happy in past wk',
                        'Spend evening with friends',
                        'Spend evening with relatives',
                        'Respondents income',
                        'Number of hours usually work a week',
                        'Job or housework',
                        'Hours of internet use on weekends',
                        'Hours of internet use on weekdays',
                        'Use twitter']

data = data[variables_of_interest]
```

```python
data.head()
```

## Rename variables

```python
data = data.rename(columns={'Year of birth': 'birth_year',
                     'Region of interview': 'region',
                     'Condition of health': 'health',
                     'General happiness': 'general_happy',
                     'How much time felt happy in past wk': 'past_week_happy',
                     'Spend evening with friends': 'see_friends',
                     'Spend evening with relatives': 'see_relatives',
                     'Respondents income': 'income',
                     'Number of hours usually work a week': 'work_hours',
                     'Job or housework': 'job_satisfaction',
                     'Hours of internet use on weekends': 'weekend_internet',
                     'Hours of internet use on weekdays': 'weekday_internet',
                     'Use twitter': 'use_twitter'})
```

```python
data.head()
```

## Create new column with total hours of internet use per week

```python
#convert to numbers, change 'don't know' answers to NaN
data = data.replace("Don't know", np.NaN)

pd.to_numeric(data['weekend_internet'])
pd.to_numeric(data['weekday_internet'])
```

```python
data['internet_per_week'] = data['weekend_internet'] * 2 + data['weekday_internet'] * 5
```

## Clean up income column 
Take middle of range for each bracket as the value

```python
#get list of all range values
all_ranges = list(data['income'].unique())
all_ranges
```

```python
#replace special cases
data['income'].replace('Refused', np.NaN, inplace = True)
data['income'].replace('$170000 or over', 170000, inplace = True)
data['income'].replace('Under $1 000', 1000, inplace = True)
```

```python
#remove special cases from the list
all_ranges = list(data['income'].unique())
all_ranges.remove(np.NaN)
all_ranges.remove(170000)
all_ranges.remove(1000)
```

```python
#make list for middle value (which will be the average) of each range
middle_values = []

#break up ranges and get middle value for each one
for value in all_ranges:
    
    #split up numbers
    rnge = value.split(' to ', 1)
        
    #get rid of whitespace
    for number in rnge:
            
        index = rnge.index(number)
            
        number = number.replace(' ', '')
        number = number.replace('$', '')
            
        rnge[int(index)] = int(number)
            
    middle_value = (rnge[0]+rnge[1])/2
        
    middle_values.append(int(middle_value))
```

```python
#replace all income ranges with the middle value in the dataframe
for value in middle_values:
    
    data['income'].replace(all_ranges[middle_values.index(value)], value, inplace = True)
```

```python
data['income']
```

## Replace birth years with people's ages in 2016

```python
for birth_year in list(data['birth_year']):
    
    #if birth_year is not recorded as NaN:
    if np.isfinite(birth_year):
    
        #replace it with what that person's age would be in 2016, the year survey was conducted
        data['birth_year'].replace(birth_year, 2016-int(birth_year), inplace = True)
```

```python
#rename birth_year column to 'age'
data = data.rename(columns={'birth_year': 'age'})
```
