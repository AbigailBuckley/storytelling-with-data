```python
import hypertools as hyp
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('sample_data.csv')
```

```python
hyp.plot(data, 'o', group=data['FT'])
```

```python
HT = data['HT']
FT = data['FT']
```

```python
HT[0]
```

```python
first_half = list(map(lambda x: list(map(int, x.split('-'))), HT))
final_game = list(map(lambda x: list(map(int, x.split('-'))), FT))
```

```python
second_half = list(map(lambda x,y: [x[0] - y[0], x[1] - y[1]], final_game, first_half))
```

```python
hyp.plot([np.array(first_half), np.array(second_half)], 'o')
```
