## World Happiness Report

Dataset: https://www.kaggle.com/unsdsn/world-happiness

This dataset contains information about 158 countries' general happiness scores. Each country's score is broken down into 6 factors:

- **Economy** (GDP per capita)
- **Family** (social support)
- **Health** (life expectancy)
- **Freedom** (freedom to make life choices)
- **Trust** (perception of government corruption)
- **Generosity**

The last column, **Dystopia Residual**, accounts for the rest of the general happiness score (not explained by the above factors) that puts that country's score above that of a fictional "dystopia"

### Import world happiness report data

```python
import pandas as pd
import numpy as np
import hypertools as hyp
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#download happiness report csv as dataframe
happiness_report = pd.read_csv('2015.csv')

#drop happiness rank column
happiness_report = happiness_report.drop('Happiness Rank', axis =1)
```

```python
happiness_report
```

### Are happiness scores clumped based on region?

```python
hyp.plot(happiness_report, 'o', group=happiness_report['Region'], legend = happiness_report['Region'].unique())
```

### What is the average happiness score by region?

```python
#for each region, get a barplot with average happiness score
sns.barplot(data=happiness_report, x='Region', y='Happiness Score')
plt.xticks(rotation=90) 
```

### Are certain factors weighted more than others?

### Do regions differ based on what happiness factor you're looking at?

```python
#break down into each happiness factor
for factor in ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual']:
    sns.barplot(data=happiness_report, x='Region', y=factor)
    plt.xticks(rotation=90) 
    plt.show()
```

```python
factors = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual']

#get mean value for each factor
means = []

for factor in factors:
    mean = np.mean(happiness_report[factor])
    means.append(mean)
    
#barplot
sns.barplot(x = factors, y = means)
plt.xticks(rotation=90) 
```
