```python
import pandas as pd
import numpy as np
import hypertools as hyp
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#download happiness report csv as dataframe
happiness_report = pd.read_csv('2015.csv')

#drop happiness rank column
happiness_report = happiness_report.drop('Happiness Rank', axis =1)
```

```python
#Economy(GDP per capita)
#Family (social support)
#Health (life expectancy)
#freedom (to make life choices)
#trust (perception of government corruption)
#generosity
```

```python
happiness_report
```

```python
hyp.plot(happiness_report, 'o', group=happiness_report['Region'], legend = happiness_report['Region'].unique())
```

```python
#for each region, get a barplot with average happiness score
sns.barplot(data=happiness_report, x='Region', y='Happiness Score')
plt.xticks(rotation=90) 
```

```python
#break down into each happiness factor
for factor in ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual']:
    sns.barplot(data=happiness_report, x='Region', y=factor)
    plt.xticks(rotation=90) 
    plt.show()
```
